```python
# Import the necessary modules and libraries
import numpy as np
from sklearn.tree import DecisionTreeRegressor
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score

# Create a random dataset
rng = np.random.RandomState(1)
X = np.sort(5 * rng.rand(80, 1), axis=0)
y = np.sin(X).ravel()
y[::5] += 3 * (0.5 - rng.rand(16))

# Fit regression model
regr_1 = DecisionTreeRegressor(max_depth=2)
regr_1.fit(X, y)


# Predict
X_test_no_split = X
y_1 = regr_1.predict(X_test_no_split)

#Split training / test data

from sklearn.model_selection import train_test_split
# TODO: Shuffle and split the data into training and testing subsets
X_train, X_test, y_train, y_test = train_test_split( X,  y, test_size=0.2)
print "Training and testing split was successful"

# Fit regression model
regr_2 = DecisionTreeRegressor(max_depth=2)
regr_2.fit(X_train, y_train)

y_2 = regr_2.predict(X_test)

# Plot the results
plt.figure()
plt.scatter(X, y, s=20, edgecolor="black", c="darkorange", label="data")
plt.plot(X_test_no_split, y_1, color="cornflowerblue", label="max_depth=2", linewidth=2)
plt.scatter(X_test, y_2, color="yellowgreen", label="max_depth=2", linewidth=2)
plt.xlabel("data")
plt.ylabel("target")
plt.title("Decision Tree Regression")
plt.legend()
plt.show()

print "R2 Score 1: {} ".format(r2_score(y, y_1))
print "R2 Score 1: {} ".format(r2_score(y_test, y_2))
```
